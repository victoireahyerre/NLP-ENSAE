# Statistiques sur les labels

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le fichier CSV
df = pd.read_csv("data\obsinfox.csv")

# Liste des labels d'annotation
labels = ['Fake News', 'Places, Dates, People', 'Facts', 'Opinions', 'Subjective',
          'Reported information', 'Sources Cited', 'False Information', 'Insinuation',
          'Exaggeration', 'Offbeat Title']

# Figure 4 : Moyenne des annotations et accord entre annotateurs
mean_scores = df[labels].mean()
agreement_scores = df.groupby("Title")[labels].mean().apply(lambda x: abs(2*x - 1)).mean()

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.barplot(x=mean_scores.index, y=mean_scores.values, ax=axes[0], color="blue")
axes[0].set_title("Moyenne des annotations par label")
axes[0].set_ylabel("Score moyen")
axes[0].tick_params(axis='x', rotation=90)

sns.barplot(x=agreement_scores.index, y=agreement_scores.values, ax=axes[1], color="green")
axes[1].set_title("Accord entre annotateurs par label")
axes[1].set_ylabel("Score d'accord moyen")
axes[1].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()

# Figure 5 : Matrice de corrélation entre les labels
corr_matrix = df[labels].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matrice de corrélation entre les labels")
plt.show()
```
