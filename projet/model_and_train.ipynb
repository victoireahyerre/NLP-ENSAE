# Entrainement et évaluation des modèles pour différents sets de labels

```python
def train_and_evaluate(label_columns, num_epochs=30, batch_size=16, learning_rate=5e-5):
    # Charger le DataFrame
    df = pd.read_csv("data/obsinfox_final.csv")
    texts = df['Article_Text'].tolist()
    labels = df[label_columns].values
    
    # Tokenisation
    tokenizer = CamembertTokenizer.from_pretrained("camembert-base")
    encoded_texts = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")
    
    # Diviser les données
    train_inputs, test_inputs, train_labels, test_labels = train_test_split(
        encoded_texts['input_ids'], labels, test_size=0.2, random_state=42
    )
    train_masks, test_masks, _, _ = train_test_split(
        encoded_texts['attention_mask'], labels, test_size=0.2, random_state=42
    )
    
    # Créer les datasets et data loaders
    train_dataset = TensorDataset(train_inputs, train_masks, torch.tensor(train_labels, dtype=torch.float32))
    test_dataset = TensorDataset(test_inputs, test_masks, torch.tensor(test_labels, dtype=torch.float32))
    train_dataloader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_dataloader = DataLoader(test_dataset, batch_size=batch_size)
    
    # Initialiser le modèle
    num_labels = len(label_columns)
    model = CamembertForSequenceClassification.from_pretrained("camembert-base", num_labels=num_labels)
    model.to(torch.device("cuda" if torch.cuda.is_available() else "cpu"))
    optimizer = torch.optim.AdamW(model.parameters(), lr=learning_rate)
    loss_fn = torch.nn.BCEWithLogitsLoss()
    
    # Entraînement
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    train_losses = []
    
    for epoch in range(num_epochs):
        model.train()
        total_loss = 0
        for batch in train_dataloader:
            input_ids, attention_mask, labels_batch = batch
            input_ids, attention_mask, labels_batch = input_ids.to(device), attention_mask.to(device), labels_batch.to(device)
            optimizer.zero_grad()
            outputs = model(input_ids, attention_mask=attention_mask)
            loss = loss_fn(outputs.logits, labels_batch)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        avg_loss = total_loss / len(train_dataloader)
        train_losses.append(avg_loss)
        print(f"Epoch {epoch+1}/{num_epochs} - Loss: {avg_loss:.4f}")
    
    # Visualisation des pertes
    plt.plot(range(1, num_epochs+1), train_losses, label='Train Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Évolution de la perte')
    plt.legend()
    plt.show()
    
    # Évaluation
    model.eval()
    total_loss = 0
    accuracies = [0] * num_labels
    
    for batch in test_dataloader:
        input_ids, attention_mask, labels_batch = batch
        input_ids, attention_mask, labels_batch = input_ids.to(device), attention_mask.to(device), labels_batch.to(device)
        with torch.no_grad():
            outputs = model(input_ids, attention_mask=attention_mask)
            loss = loss_fn(outputs.logits, labels_batch)
            total_loss += loss.item()
            predictions = torch.sigmoid(outputs.logits) > 0.5
            labels_batch = labels_batch > 0.5
            for i in range(num_labels):
                accuracies[i] += (predictions[:, i] == labels_batch[:, i]).sum().item()
    
    avg_loss = total_loss / len(test_dataloader)
    print(f"Test Loss: {avg_loss:.4f}")
    for i, col in enumerate(label_columns):
        print(f"Test Accuracy - {col}: {accuracies[i] / len(test_labels):.4f}")

# Utilisation avec différents ensembles de labels
train_and_evaluate(['Fake News'])
train_and_evaluate(['Fake News', 'Opinions', 'Subjective', 'Sources Cited'])
train_and_evaluate([col for col in df.columns if col != 'Article_Text'])
```
